#### independent and dependent data


````python
x = df.iloc[:,:-1]
y = df.iloc[:,-1]
````

````python
y.head()
````

````python
x.head()
````

````python
# spliting
from sklearn.model_selection import train_test_split # type: ignore
x_train,x_test,y_train ,y_test = train_test_split(x,y,test_size=0.2,random_state=40)
````

````python
x_train.head()
````

````python
x_test.head()
````

````python
# Standrdization
from sklearn.preprocessing import StandardScaler # type: ignore

scaler = StandardScaler()

x_train = scaler.fit_transform(x_train)

x_test = scaler.transform(x_test)
````

````python
x_train,x_test
````

### Model Training

````python
from sklearn.linear_model import LinearRegression

model = LinearRegression()

model.fit(x_train,y_train)
````

````python
# printing the coefficent and intercept

print(model.coef_) # no. of indep variables = no. of coeff
````

````python
print(model.intercept_)
````

````python
# on which parameters model has trained
model.get_params()
````

````python
## pridicting
reg_pre = model.predict(x_test)
reg_pre
````

````python
# plot a scatter of predictions
plt.scatter(y_test,reg_pre)
````

````python
## residuals (error)
residuals = y_test-reg_pre
residuals
````

````python
sns.displot(residuals,kind='kde')
````

````python
plt.scatter(reg_pre,residuals)
````

````python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
````

````python
print('MAE',mean_absolute_error(y_test,reg_pre))
print('MSE',mean_squared_error(y_test,reg_pre))
print('Rmse',np.sqrt(mean_squared_error(y_test,reg_pre)))
````

````python
# r2_score
from sklearn.metrics import r2_score
print('r2',r2_score(y_test,reg_pre))
````
